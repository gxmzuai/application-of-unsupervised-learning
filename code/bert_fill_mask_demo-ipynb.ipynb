```python
from transformers import pipeline, AutoModelForMaskedLM, BertTokenizer
```

```python
# 确保您已经在运行之前挂载了Google Drive，并且模型已经保存到指定的路径
from google.colab import drive
drive.mount('/content/drive')
```

```python
# 加载保存的模型和tokenizer
model_path = "/content/drive/My Drive/Colab Notebooks/imdb_sentiment_analysis_with_bert/trained_model"  # 更新为您的模型保存路径
model = AutoModelForMaskedLM.from_pretrained(model_path)
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
# 将模型和tokenizer传递给pipeline
fill_mask = pipeline(
    "fill-mask",
    model=model,
    tokenizer=tokenizer
)
```

```python
# 输入句子并执行填空任务
sentences = [
    "The new movie is so [MASK].",
    "Nothing is more [MASK] than discovering the truth.",
    "I think therefore I [MASK]."
]
```

```python
# 运行填空任务
for sentence in sentences:
    result = fill_mask(sentence)
    print(f"Input: {sentence}")
    for res in result:
        print(f"Predicted: {res['sequence']}, score: {res['score']:.4f}")
    print("\n" + "-"*50 + "\n")
```
